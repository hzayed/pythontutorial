```python
import pandas as pd
import numpy as np
pstore = pd.read_csv("covid_19_india.csv")
pstore.head(10)
```

```python
#Create a distribution plot for sno
sns.distplot(pstore.Sno)
plt.show()

#Change the number of bins
sns.distplot(df.Sno, bins=22, kde = False)
plt.show()

#Create a distribution plot for rating to change the title of the figure,colours,size of font
sns.distplot(pstore.Sno, bins=20, color="g")
plt.title("Distribution of app ratings", fontsize=20, color = 'red')
plt.show()

#Adding dark background to the graph
plt.style.use("dark_background")
plt.show()

#بدي انظف الداتا
# #Remove the rows with values which are less represented 
pstore = pstore[~pstore['Sno'].isin([100])]

#Resetting the index
pstore.reset_index(inplace=True, drop=True)

#Analyzing the Content Rating column again
pstore['Sno'].value_counts()
```

```python
#Plotting a bar chart
plt.figure(figsize=[9,7])
pstore['State/UnionTerritory'].value_counts().plot.barh()
plt.show()
```

```python
plt.scatter(df.Date,df.Time)# time and date are columns
plt.show()
```

```python
# Plotting the same thing now using a jointplot(seaborn)
sns.jointplot(pstore.Date, pstore.Time)
plt.show()
```

```python
# Plotting the same thing now using a jointplot(Regression Plot)
sns.jointplot(pstore.Date, pstore.Time, kind = "reg")
plt.show()

```

```python
# Plotting the same thing now using a jointplot(pair plot)
sns.pairplot(pstore[['Sno', 'Date', 'Time','State/UnionTerritory']])
plt.show()
```
